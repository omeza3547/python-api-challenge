```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json

# Import API key
from api_keys import weather_api_key as weather_key
```

```python
# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
city= "mataura"
url= f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={weather_key}&units=imperial"
response= requests.get(url).json()
```

```python
temps= []
cloudiness= []
humidity= []
max_temp= []
wind_sp= []
longe= []
lat= []
country= []
city_name= []
date= []


for city in cities:
    print("--------------")
    print(city)
    url= f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={weather_key}&units=imperial"

    response= requests.get(url).json()
    try:
    
        
        #collect data and add to list.
        temps.append(response["main"]["temp"])
        cloudiness.append(response["clouds"]["all"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        wind_sp.append(response["wind"]["speed"])
        longe.append(response["coord"]["lon"])
        lat.append(response["coord"]["lat"])
        country.append(response["sys"]["country"])
        city_name.append(response["name"])
        date.append(response["dt"])
        
    except:
        print("City Not found")
        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
global_weather_df= pd.DataFrame({"City": city_name, "Country": country,"Latitude": lat,
                                 "Longitude": longe, "Date": date, "Temperature": temps,
                                "Max Temperature": max_temp, "Humidity": humidity,
                                "Wind Speed": wind_sp, "Cloudiness": cloudiness})
global_weather_df.to_csv("../output_data/global_weather.csv", index=False)
global_weather_df.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
nx_axis = north_hem_weather["Latitude"]
maxt_axis = north_hem_weather["Max Temperature"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(nx_axis, maxt_axis)
regress_values = nx_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(nx_axis,maxt_axis)
plt.plot(nx_axis,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=14,color="red")
plt.xlabel('Latitude')
plt.ylabel(f"Max Temperature(F)")
plt.title(f"Northern Hemisphere - Max Temp vs. Latitude ({today})")
plt.grid(linewidth=.5)
plt.show()
plt.savefig("../output_data/nhem_lat_maxt_linr.png")
#north_hem_weather.plot(kind="scatter", x= "Latitude", y="Max Temperature")
```
